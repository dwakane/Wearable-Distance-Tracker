Thread mill velocities readings in Miles/Hour

```python
walk_v_mph = [0,1,2,3,4]
run_v_mph = [3,4,4,6,8,8]

walk_measurements = [stand0,walk1,walk2,walk3,walk4]
run_measurements = [run3,run4_0,run4_1,run6,run8_0,run8_1]

walk_mean_abs = [np.absolute(x).mean() for x in walk_mean_abs]
run_mean_abs = [np.absolute(x).mean() for x in run_measurements]

walk_v_mph = np.array(walk_v_mph)
run_v_mph = np.array(run_v_mph)

walk_mean_abs = np.array(walk_mean_abs)
run_mean_abs = np.array(run_mean_abs)

mean_abs = np.append(walk_mean_abs,run_mean_abs)
v_mph = np.append(walk_v_mph,run_v_mph)

linear_fit = np.polyfit(mdps_digit*mean_abs,v_mph,1)
linear = np.poly1d(linear_fit)
linear_x = np.linspace(mdps_digit*mean_abs.min(),mdps_digit*mean_abs.max(),2)
correlation = np.corrcoef(mean_abs,v_mph)
print("Correleation:")
print(correlation)
print("\nSlope in MPH:")
print(np.polyfit(mean_abs,v_mph,1)[0])
```

Strong Possitive Correlation between Velocity and Mean ABS Readings

```python
scaling = np.polyfit(mean_abs,v_mph,1)[0]
mph_to_fps = 1.466667
scale_fps_digits = scaling**mph_to_fps
print(scale_fps_digits , "fps/digits")
```
